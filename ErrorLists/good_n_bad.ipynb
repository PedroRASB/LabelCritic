```python
th=0.5
def get_bad_results(results,th=0.5):
    organs=results.keys().to_list()[1:]

    nnUNetBad={}
    for organ in organs:
        nnUNetBad[organ] = {}

    for index, row in results.iterrows():
        for key in names.keys():
            if row['name'] in names[key]:
                if 'spleen' in key:
                    if row['spleen']<th:
                        nnUNetBad['spleen'][row['name']] = row['spleen']
                elif 'stomach' in key:
                    if row['stomach']<th:
                        nnUNetBad['stomach'][row['name']] = row['stomach']
                else:
                    for organ in organs:
                        if row[organ]<th:
                            nnUNetBad[organ][row['name']] = row.to_dict()[organ]
    print('Threshold:', th)                        
    print('Number of bad results:', sum([len(nnUNetBad[organ]) for organ in nnUNetBad.keys()]))
    for organ in organs:
        print(f'Number of bad {organ}:', len(nnUNetBad[organ]))
    return nnUNetBad
```

```python
nnUNetBad05=get_bad_results(results=results_nn,th=0.5)
nnUNetBad06=get_bad_results(results=results_nn,th=0.6)
nnUNetBad07=get_bad_results(results=results_nn,th=0.7)
nnUNetBad08=get_bad_results(results=results_nn,th=0.8)
nnUNetBad09=get_bad_results(results=results_nn,th=0.9)
```

# Save Pairs

```python
organs=results_nn.keys().to_list()[1:]

bad_old_all={'DSC<=0.5':get_bad_results(results=comparison_results,th=0.5),
             'DSC<=0.6':get_bad_results(results=comparison_results,th=0.6),
             'DSC<=0.7':get_bad_results(results=comparison_results,th=0.7),
             'DSC<=0.8':get_bad_results(results=comparison_results,th=0.8),
             'DSC<=0.9':get_bad_results(results=comparison_results,th=0.9)}
bad_nn_all={'DSC<=0.5':get_bad_results(results=results_nn,th=0.5),
        'DSC<=0.6':get_bad_results(results=results_nn,th=0.6),
        'DSC<=0.7':get_bad_results(results=results_nn,th=0.7),
        'DSC<=0.8':get_bad_results(results=results_nn,th=0.8),
        'DSC<=0.9':get_bad_results(results=results_nn,th=0.9)}




bad_old_labels={'DSC<=0.5':{},
                '0.5<DSC<=0.6':{},
                '0.6<DSC<=0.7':{},
                '0.7<DSC<=0.8':{},
                '0.8<DSC<=0.9':{},
                }
bad_nnUNet={'DSC<=0.5':{},
            '0.5<DSC<=0.6':{},
            '0.6<DSC<=0.7':{},
            '0.7<DSC<=0.8':{},
            '0.8<DSC<=0.9':{},
            }
th_name=None
for th in [0.5,0.6,0.7,0.8,0.9]:
    old_th_name=th_name
    if th==0.5:
        th_name='DSC<=0.5'
    else:
        th_name=f'{str(np.round(th-0.1,1))}<DSC<={str(th)}'
    for key in organs:
        if th==0.5:
            if key in bad_old_all[th_name].keys():
                bad_old_labels[th_name][key] = list(bad_old_all[th_name][key].keys())
            bad_nnUNet[th_name][key] = list(bad_nn_all[th_name][key].keys())
        else:
            #print(bad_old_labels.keys())
            if old_th_name=='DSC<=0.5':
                old='DSC<=0.5'
            else:
                old=old_th_name[4:]
            #print('th_name: ',th_name)
            if key in bad_old_all[th_name[4:]].keys():
                bad_old_labels[th_name][key] = [item for item in list(bad_old_all[th_name[4:]][key].keys())\
                                                if item not in list(bad_old_all[old][key].keys())]
            bad_nnUNet[th_name][key] = [item for item in list(bad_nn_all[th_name[4:]][key].keys()) \
                                        if item not in list(bad_nn_all[old][key].keys())]

        
```
